```python
df['residence_type'].replace({1:'urban',2:'rural'})
```

```python
df['religion'].value_counts()
```

```python
df['religion']=df['religion'].replace({
    '01': 'Hindu',
    '02': 'Muslim',
    '03': 'Christian',
    '04': 'Sikh',
    '05': 'Buddhist',
    '06': 'Jain',
    '07': 'Jewish',
    '08': 'Parsi',
    '09': 'No religion',
    '96': 'Other'
})
```

```python
df['education_level'].value_counts()
```

```python
df['education_level']=df['education_level'].replace({
    0: 'No education',
    1: 'Primary (1–5)',
    2: 'Middle (6–8)',
    3: 'Secondary (9–10)',
    4: 'Higher secondary (11–12)',
    5: 'Higher education'
})
```

```python
df['resp_height']=df['resp_height']/10
```

```python
df['heamoglobin']=df['hemoglobin']/10
```

```python
df['heamoglobin_adj']=df['hemoglobin_adj']/10
```

```python
df['anemia_level'].value_counts()
```

```python
df['anemia_level']=df['anemia_level'].replace({
    1.0: 'No anemia',
    2.0: 'Mild anemia',
    3.0: 'Moderate anemia',
    4.0: 'Severe anemia'

})
```

```python
df['caste'].value_counts()
```

```python
df['caste']=df['caste'].replace({
    1.0: 'Scheduled Caste (SC)',
    2.0: 'Scheduled Tribe (ST)',
    3.0: 'Other Backward Class (OBC)',
    4.0: 'General',
    8.0: 'Don’t know'
})
```

```python
df['marital_status'].value_counts()
```

```python
df['marital_status']=df['marital_status'].replace({
    0: 'Never married',
    1: 'Currently married',
    2: 'Married, gauna not performed / Formerly married',
    3: 'Widowed',
    4: 'Divorced',
    5: 'Separated',
    6: 'Deserted'
})
```
